```python
import pandas as pd 
melb_data=pd.read_csv("C:\SkillFactory\PYTHON-12. Продвинутые методы работы с данными в Pandas\data\melb_data_fe.csv", sep=",")
melb_dt=melb_data.copy()
melb_dt["Date"]=pd.to_datetime(melb_dt["Date"])
dt1="2017-05-01"
dt2="2017-09-01"
mask=(dt1<=melb_data["Date"])&(melb_data["Date"]<=dt2)
melb_dt[mask].groupby("SellerG")["Price"].sum().sort_values()
max_unit=melb_dt.pivot_table(
    values="Price",
    index="SellerG",
    columns="Type",
    aggfunc="median"
)
mask=max_unit["unit"].max()
print(max_unit[max_unit["unit"]==mask].index[0])
```

```python
import pandas as pd 
import re
ratings_movies=pd.read_csv("./data/ratings_movies.csv", sep=",")
ratings_movies_copy=ratings_movies.copy()

def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
ratings_movies_copy["year_release"]=ratings_movies_copy["title"].apply(get_year_release)
ratings_movies_copy=ratings_movies_copy[(ratings_movies_copy['year_release']==2018)]
ratings_movies_copy_new=ratings_movies_copy.groupby("genres")["rating"].agg(["mean", "count"])

ratings_movies_copy_new=ratings_movies_copy_new[(ratings_movies_copy_new["count"]>10)]["mean"].sort_values(ascending=False)
display(ratings_movies_copy_new)
```
