```python
# создаем датафрейм считав в него файл
df=pd.read_json("C:\SkillFactory\учебный материал\data\\recipes.json")
# создаем копию
df_copy=df.copy()
display(df_copy)
# создаем пустой список уникальных ингредиентов
all_ingredients=[]
# при помощи цикла проходимся по все рецептам и добавляем уникальные значение в список
for i in df_copy["ingredients"]:
    for y in i:
        if y not in all_ingredients:
            all_ingredients.append(y)
# выводим кол-во уникальных ингредиентов
print(len(all_ingredients))
# при помощи цикла проходимся по списку уникальных ингредиентов
for i in all_ingredients:
    # применяем лямбду функцию и создаем новые признаки с названием ингредиентов
    df_copy[i]=df_copy["ingredients"].apply(lambda x: 1 if i in x else 0)
# изменим значения признака ingredients и укажим кол-во ингредиентов
df_copy["ingredients"]=df_copy["ingredients"].apply(len)
display(df_copy)
df.to_csv('recipes.csv', index = False)
    
```

```python
# создаем список уникальных id 
ids = list(df_copy['id'].unique())
# Создаем список уникальных значений ингредиентов
ingredients = list(df.columns)[3:] 
df = pd.read_csv('recipes.csv') # Читаем содержимое 
ingredients = list(df.columns)[3:] # Создаем список 
# Определяем имя функции и аргументы
def make_list(row): 
  # Создаем пустой список ингредиентов проверяемого блюда
  ingredient_list=[] 
  # Перебираем ингредиенты из реестра
  for ingredient in ingredients: 
    # Если ингредиент входит в состав текущего блюда
    if row[ingredient].item()==1: 
      # Добавляем его в список ингредиентов текущего блюда
      ingredient_list.append(ingredient) 
  # Возвращаем список ингредиентов
  return ingredient_list 
# Создаём пустой список для хранения итоговой структуры
new_recipes = [] 
# Организуем цикл с параметром current_id
for current_id in ids: 
    # Получаем значение соответствующей кухни, применив фильтр по текущему значению параметра цикла к DataFrame;
    cuisine = df_copy[df_copy['id'] == current_id]['cuisine'].iloc[0] 
    # Получаем перечень ингредиентов, входящих в состав текущего блюда
    current_ingredients = make_list(df_copy[df_copy['id'] == current_id]) 
    # Создаём текущий словарь
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients} 
    # Добавляем созданный словарь к списку
    new_recipes.append(current_recipe)
new_recipes = json.dumps(new_recipes) 

# Откроем файл new_recipes.json для записи
with open("C:\SkillFactory\учебный материал\data\\new_recipes.json", "w") as write_file: 
    # Записываем содержимое подготовленные данные в файл
    write_file.write(new_recipes)
```

```python
# импортируем библиотеку для работы с xml
import xml.etree.ElementTree as ET
# заносим в переменную файл xml
tree=ET.parse("C:\SkillFactory\учебный материал\data\menu.xml")
# добавляем в переменную корень файла
root=tree.getroot()
display(root)
# посмотрим потомков корневого узла
display(list(root))
# выводим списко потомков второго блюда(второго потомка корневого узла)
display(list(root[1]))
# выведем атрибуты первого блюда
display(root[0].attrib)
# выведем значение price первого блюда и прочитаем его с помощью функции text
display(root[0][0].text)
# выведем наименование тега по индексу
display(root[0][2].tag)
# циклом проходимся по потомкам
for dish in root:
    # циклом проходимся по атрибутам потомков
    for param in dish:
        # выводим на экран название блюда и его атрибутов со значениями
        print(dish.attrib["name"], param.tag, param.text)
    print()

import pandas as pd
# создаем список с названием столбцов
column_names = ['name', 'price', 'weight', 'class']
# создаем пустой список
df_list = []
# цклом проходимся по потомкам корня
for dish in root:
    # в переменную заносим название блюда и значений его атрибутов
    row = [dish.attrib['name'], dish[0].text, dish[1].text, dish[2].text]
    # добавляем строку выше в список
    df_list.append(row)
# создаем датафрейм и заносим в него список со значениями и названием колонок
df = pd.DataFrame(df_list, columns=column_names)
display(df)
```
