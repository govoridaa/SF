```python
# пытаемся считать файл не зная кодировки
data=pd.read_csv("C:\SkillFactory\учебный материал\data\ErrorEnCoding.csv", encoding_errors="replace", header=None)
display(data)
# определяем кодировку
# импортиуем субмодуль
from chardet.universaldetector import UniversalDetector
# создаем детектор
detector=UniversalDetector()

with open("C:\SkillFactory\учебный материал\data\ErrorEnCoding.csv", "rb") as fh:
    for line in fh:
        detector.feed(line)
        if detector.done:
            break

print(detector.close())

# получим результат {'encoding': 'KOI8-R', 'confidence': 0.8773902118791048, 'language': 'Russian'}
# который говорит нам, что с вероятностью 87% что используется кодировка 'KOI8-R'
# считываем файл с указанием явной кодировки
data=pd.read_csv("C:\SkillFactory\учебный материал\data\ErrorEnCoding.csv", encoding="KOI8-R", header=None)
display(data)
```
