```python
import pandas as pd
wine=pd.read_csv("C:\SkillFactory\учебный материал\data\wine_cleared.csv")
# Создадим новый признак price_round, означающий округлённую до целого числа цену за бутылку вина
wine['price_round']=wine['price'].round().astype(int)
# переменная с реуглярным выражением \d щзначет поиск чисел в диапозоне от 0-9 и {4} четырехзначное
regex='\d{4}'
# при помощи findall - находим все вхождения, а при помощи str.get(0) находим
#  первое значение из списка всех вхождений и добавлем в колонку
wine['year']=wine['title'].str.findall(regex).str.get(0)
# добавим еще признак если вино произведенно в сша то 1 иной вариант 0
wine['is_usa']=wine['country'].apply(lambda x: 1 if x=='US' else 0)
# добавим еще признак если вино произведенно в сша то 1 иной вариант 0
wine['is_france']=wine['country'].apply(lambda x: 1 if x=='France' else 0)
# добавим еще признак если вино произведенно в сша то 1 иной вариант 0
wine['is_italy']=wine['country'].apply(lambda x: 1 if x=='Italy' else 0)
# при помощи to_datetime мы преобразовали в датуБ а errors='coerce' преобразовали значение в
#  число а нечисловое значение в Nan
wine['year'] = pd.to_datetime(wine['year'], errors='coerce')
wine['old_wine']=wine['year'].apply(lambda x: 1 if x.year<2010 else 0)
# регулярное выражение \( означет начальный символ. ( группа захвата начало. '.' означет один
#  произвольный символ. '*' означет произвольное кол-во повторений одного символа. ? означет стоп
#  когда встретит конечный символ. ) группа захвата конец. \) конечный символ
regex2='\((.*?)\)'
wine['locality']=wine['title'].str.findall(regex2).str.get(0)
display(wine)
```

```python
import category_encoders as ce
import pandas as pd
# инициализируем информацию об одежде
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list, columns = ['size',  'type'])
ord_encoder = ce.OrdinalEncoder(mapping=[{
	'col': 'size',
	'mapping': {'xxs': 1, 'xs': 2, 's': 3, 
                'm': 4, 'l': 5, 'xxl': 6}
}])
data_bin = ord_encoder.fit_transform(clothing[['size']])
clothing = pd.concat([clothing, data_bin], axis=1)
clothing
```
