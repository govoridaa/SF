Ответ:  Отток клиентов составляет чуть больше 20%. Из этого можно сделать, что каждый пятый клиент банка прекращает пользовтаься его услугами. Есть необходимость принять меры по снижению данного показателя хотябы в два раза.

9.8. Как влияет наличие статуса активного клиента на отток клиентов? Постройте диаграмму, иллюстрирующую это. Что бы вы предложили банку, чтобы уменьшить отток клиентов среди неактивных?

Ответ: По первой диаграмме видно, что больше активных клиентов среди лояльных. И полная противоположность среди ушедших, это видно по второй диаграмме большая часть пользуется одной услугой банка и резкий спад на доп услуги. Это говорит о том, что неактивные клиенты со временем уходят из банка. По этому рекомендация консультативный звонок по услугам и акциям банка, для выяснения причин неактивности. Это может быть связанно с тем, что клиент во все не знает об услугах или акциях банка. 

```python
import seaborn  as sns
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
churn=pd.read_csv(".\data\churn.csv")
churn_copy=churn.copy()
display(churn_copy)
choropleth_data = churn_copy.groupby(['Geography'])['Exited'].mean().reset_index()
IsActiveMember_data = churn_copy.groupby(['Geography'])['IsActiveMember'].mean().reset_index()
fig=px.choropleth(
    data_frame=choropleth_data,
    locations="Geography",
    locationmode = "country names", 
    color="Exited",
    range_color=[0,1],
    title="Доля потерянных клиентов по странам",
    width=800,
    height=500,
    color_continuous_scale='Reds'
)
fig1=px.choropleth(
    data_frame=IsActiveMember_data,
    locations="Geography",
    locationmode = "country names", 
    color="IsActiveMember",
    range_color=[0,1],
    title="Доля активных клиентов по странам",
    width=800,
    height=500,
    color_continuous_scale='Reds'
)
fig.show()
fig1.show()

```

9.9. В какой стране доля ушедших клиентов больше? Постройте тепловую картограмму, которая покажет это соотношение на карте мира. Предположите, с чем это может быть связано.

Ответ: По данному графику видно, что кол-во ушедших клиентов практически в два раза выше чем в других странах. Исходя из того, что в германии больше не активных клиентов(что видно на второй диаграмме), то решить проблему поможет повышение активности клиентов, акции, доп услуги. 

```python
import seaborn  as sns
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
churn=pd.read_csv(".\data\churn.csv")
churn_copy=churn.copy()
def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score >= 851:
        return "Top"
    elif credit_score < 300:
        return "Deep"
churn_copy["CreditScoreCat"]=churn_copy["CreditScore"].apply(get_credit_score_cat)
pivot=churn_copy.pivot_table(
    "Exited",
    index="CreditScoreCat",
    columns="Tenure",
    aggfunc="mean"
)
heatmap=sns.heatmap(
    data=pivot,
    annot=True,
    cmap='YlGnBu'
)
heatmap.set_title("Ушедшие клиенты в разрезе кредитного рейтинга", fontsize=16)
display(pivot)
```
